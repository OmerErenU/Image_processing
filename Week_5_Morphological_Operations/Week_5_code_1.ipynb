```python
new_img = cv2.imread("./resim_2.jpg",0)
```

```python
white_nose= np.random.randint(0,2,size=new_img.shape[:2])
white_nose = white_nose * 250
```

```python
plt.imshow(white_nose, cmap="gray")
```

```python
my_noised_img = new_img + white_nose
plt.imshow(my_noised_img ,cmap="gray")
```

```python
my_openninb_img = cv2.morphologyEx(my_noised_img.astype(np.float32),cv2.MORPH_OPEN , my_kernel)
```

```python
fig = plt.figure(figsize=(10,7))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.imshow(new_img, cmap="gray")
plt.axis("off")
plt.title("original image")



fig.add_subplot(rows,columns,2)
plt.imshow(my_noised_img, cmap="gray")
plt.axis("off")
plt.title("Noised İmage")

fig.add_subplot(rows,columns,3)
plt.imshow(my_openninb_img, cmap="gray")
plt.axis("off")
plt.title("openningİmage")




```

```python
black_nose= np.random.randint(0,2,size=new_img.shape[:2])
black_nose = black_nose * -255
```

```python
plt.imshow(black_nose , cmap="gray")
```

```python
my_black_noise_img = new_img + black_nose
plt.imshow(my_black_noise_img , cmap="gray")
```

```python
my_black_noise_img[my_black_noise_img <= -245] = 0 
```

```python
my_closing_img = cv2.morphologyEx(my_black_noise_img.astype(np.float32),cv2.MORPH_CLOSE , my_kernel)
```

```python
fig = plt.figure(figsize=(20,12))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.imshow(new_img, cmap="gray")
plt.axis("off")
plt.title("original image")



fig.add_subplot(rows,columns,2)
plt.imshow(my_black_noise_img, cmap="gray")
plt.axis("off")
plt.title(" black Noised İmage")

fig.add_subplot(rows,columns,3)
plt.imshow(my_closing_img, cmap="gray")
plt.axis("off")
plt.title(" closing openningİmage")

```

```python
my_gradient_img_1 = cv2.morphologyEx(my_original_img , cv2.MORPH_GRADIENT,my_kernel)
my_gradient_img_2 = cv2.morphologyEx(new_img , cv2.MORPH_GRADIENT,my_kernel)
```

```python
fig = plt.figure(figsize=(20,12))
rows = 2
columns = 2

fig.add_subplot(rows,columns,1)
plt.imshow(my_original_img, cmap="gray")
plt.axis("off")
plt.title("original img ")

fig.add_subplot(rows,columns,2)
plt.imshow(my_gradient_img_1, cmap="gray")
plt.axis("off")
plt.title(" closing openningİmage")

fig.add_subplot(rows,columns,3)
plt.imshow(new_img, cmap="gray")
plt.axis("off")
plt.title("new image")

fig.add_subplot(rows,columns,4)
plt.imshow(my_gradient_img_2, cmap="gray")
plt.axis("off")
plt.title(" Gradient İmage")


```
