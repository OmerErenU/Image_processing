```python
import cv2 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img_original = cv2.imread("./pappers.jpg",1)
```

```python
plt.imshow(img_original);
```

```python
img_rgb = cv2.cvtColor(img_original , cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_original , cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(10, 7)) 
  
# setting values to rows and column variables 
rows = 1
columns = 3
  
# reading images 

  
# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1)
plt.imshow(img_original) 
plt.axis('off') 
plt.title("original image") 

fig.add_subplot(rows, columns,2)
plt.imshow(img_rgb) 
plt.axis('off') 
plt.title("RGB") 

fig.add_subplot(rows, columns,3)
plt.imshow(img_gray , cmap="gray") 
plt.axis('off') 
plt.title("GRAY") 
```

```python
img_new_1 = cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
img_new_RGB_2 = cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
```

```python
# RGB Yİ HSV YE DÖNÜŞTÜR
img_hsv_green =cv2.cvtColor(img_original , cv2.COLOR_BGR2HSV)

# YEŞİL İÇİN hsv ARALIĞINI BELİRLEYİN

lower_green = np.array([20,100,90]) # yeşil için HSV  sınırı
upper_green = np.array([130,210,130]) 

mask_green = cv2.inRange(img_hsv_green , lower_green , upper_green)
```

```python
fig = plt.figure(figsize=(10, 7)) 
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(img_new_1) 
plt.axis('off') 
plt.title("rgb") 

fig.add_subplot(rows, columns,2)
plt.imshow(mask_green , cmap="gray") 
plt.axis('off') 
plt.title("RGB İNDEXES") ;
```

```python
yellow = np.full_like(img_new_1, [255,255,0])

img_new_1[mask_green != 0] = yellow[mask_green !=0]
```

```python
plt.imshow(img_new_1);
```

```python
#Kodlarla renk oluşturma 
height,width = 100 , 100

#kırmızı renk oluşturma
img_red = np.zeros((height,width,3), dtype=np.uint8)
img_red[:,:,0] = 255

#yeşil renk oluşturma
img_green = np.zeros((height,width,3), dtype=np.uint8)
img_green[:,:,1] = 255

#mavi  renk oluşturma
img_blue= np.zeros((height,width,3), dtype=np.uint8)
img_blue[:,:,2] = 255

#sarı renk oluşturma
img_yellow= np.zeros((height,width,3), dtype=np.uint8)
img_yellow[:,:,:2] = 255
```

```python
img_red.shape
```

```python
fig = plt.figure(figsize=(10, 7)) 
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(img_red) 
plt.axis('off') 
plt.title("red") 

fig.add_subplot(rows, columns,2)
plt.imshow(img_green) 
plt.axis('off') 
plt.title("green") 

fig.add_subplot(rows, columns,3)
plt.imshow(img_blue ) 
plt.axis('off') 
plt.title("blue") ;

fig.add_subplot(rows, columns, 4)
plt.imshow(img_yellow) 
plt.axis('off') 
plt.title("yellow")
```

```python
img_black= np.zeros((height,width,3), dtype=np.uint8)


img_white = np.ones((height,width,3), dtype=np.uint8)*255
```

```python
fig = plt.figure(figsize=(10, 7),facecolor='purple') 
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(img_white) 
plt.axis('off') 
plt.title("white") 

fig.add_subplot(rows, columns,2)
plt.imshow(img_black) 
plt.axis('off') 
plt.title("black") ;
```
